```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
import seaborn as sns

from glob import glob
```

```python
plt.rc('font', size=16)
plt.rc('axes', titlesize=16)
plt.rc('axes', labelsize=16)
plt.rc('xtick', labelsize=16)
plt.rc('ytick', labelsize=16)
plt.rc('legend', fontsize=16)
plt.rc('figure', titlesize=16) 
```

```python
def get_mean(path):
    files = glob(path)

    df = pd.DataFrame()

    for file_ in files:
        df = df.append(pd.read_csv(file_))
    
    return df["f1_score"].mean()
```

```python
def get_whole_dataframe(path, approach, model):
    files = glob(path)

    df = pd.DataFrame()

    for file_ in files:
        temp_df = pd.read_csv(file_)
        temp_df["approach"] = approach
        temp_df["model"] = model
        df = df.append(temp_df)
    
    return df
```

```python
all_df = pd.DataFrame()
all_df = all_df.append(get_whole_dataframe("../results/cross_species/efaecium_model_on_kpneumoniae/*plasmid.csv", "plASgraph", "K. pneumoniae-(E. faecium model)-plASgraph"))
all_df = all_df.append(get_whole_dataframe("../results/mlplasmids_cross_species/efaecium_model_on_kpneumoniae/*plasmid.csv", "mlplasmids", "K. pneumoniae-(E. faecium model)-mlplasmids"))
all_df = all_df.append(get_whole_dataframe("../results/cross_species/efaecium_model_on_ecoli/*plasmid.csv", "plASgraph", "E. coli-(E. faecium model)-plASgraph"))
all_df = all_df.append(get_whole_dataframe("../results/mlplasmids_cross_species/efaecium_model_on_ecoli/*plasmid.csv", "mlplasmids", "E. coli-(E. faecium model)-mlplasmids"))
all_df = all_df.append(get_whole_dataframe("../results/cross_species/ecoli_model_on_kpneumoniae/*plasmid.csv", "plASgraph", "K. pneumoniae-(E. coli model)-plASgraph"))
all_df = all_df.append(get_whole_dataframe("../results/mlplasmids_cross_species/ecoli_model_on_kpneumoniae/*plasmid.csv", "mlplasmids", "K. pneumoniae-(E. coli model)-mlplasmids"))
all_df = all_df.append(get_whole_dataframe("../results/cross_species/kpneumoniae_model_on_ecoli/*plasmid.csv", "plASgraph", "E. coli-(K. pneumoniae model)-plASgraph"))
all_df = all_df.append(get_whole_dataframe("../results/mlplasmids_cross_species/kpneumoniae_model_on_ecoli/*plasmid.csv", "mlplasmids", "E. coli-(K. pneumoniae model)-mlplasmids"))
```

```python
all_df.to_csv("../results/predictions_comparison/cross_species_comparison_plasmid.csv", index=False)
```

```python
plt.figure(figsize=(12,6))

p = sns.boxplot(x="model", y="f1_score", data=all_df, linewidth=1, color="white", showfliers = False)
p = sns.stripplot(x="model", y="f1_score", data=all_df, hue="approach", palette=["darkslategrey", "cadetblue"])

p.set_xticklabels(p.get_xticklabels(), rotation=40)
plt.ylabel("F1-score")

p.legend_.set_title(None)
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))

plt.savefig("../results/figures/cross_species_f1_comparison_plasmid.png", dpi=600, format="png", bbox_inches="tight")
plt.show()
plt.clf()
plt.close()
```
